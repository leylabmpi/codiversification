# trait-db

## General metadata

```r
# general genome metadata
db = genomes %>%
    select.(Name, species, ncbi_taxid, strain, dl_species, Fasta, 
            Completeness, Contamination, Strain.heterogeneity, Genome.size..bp.,
            GC, X..contigs, N50..contigs., Coding.density, X..predicted.genes) %>%
    rename.('Genome' = Name,
            'Genome_size__bp' = Genome.size..bp.,
            'N50__contigs' = N50..contigs.,
            'Num_predicted_genes' = X..predicted.genes,
            'Num_contigs' = X..contigs,
            'Strain_heterogeneity' = Strain.heterogeneity,
            'Coding_density' = Coding.density)
db
```

```r
# used for downstream processing
meta = db %>%
    distinct.(Genome, species) %>%
    rename.('Sample' = Genome)
meta
```

## Phenotypes

### Traitar

```r
# loading
F = file.path(llg_dir, 'phenotype', 'traitar', 'predictions_flat_single-votes_combined.tsv.gz')
traits = Fread(F, fill=TRUE, tmpdir=tmp_dir) %>%
    filter.(phenotype_model == 'phypat+PGL') %>%
    select.(-gene_caller, -phenotype_model) %>%
    complete.(genome, phenotype, fill=list(prediction_score='0')) %>%
    mutate.(prediction_score = as.Num(prediction_score))
traits$genome %>% unique_n('genomes')
traits$prediction_score %>% as.character %>% table
traits
```

```r
# adding metadata & formatting
overlap(traits$genome, meta$Sample)
traits = traits %>%
    inner_join.(meta %>% select.(Sample, species),
                c('genome'='Sample')) %>%
    pivot_wider.(names_from=phenotype, values_from=prediction_score) %>%
    mutate.(across.(where(is.numeric), ~replace_na.(., 0)))
traits
```

```r
# adding to db
db = db %>% 
    inner_join.(traits, c('Genome'='genome', 'species'))
traits = NULL; gc()
db
```

## Annotations

### eggnog-mapper

```r
# emapper results
F = file.path(llg_dir, 'annotate', 'eggnog-mapper', 'prokka', 'annotations.tsv.gz')
emap = Fread(F, fill=TRUE, tmpdir=tmp_dir) %>%
    select.(-eggNOG_OGs) %>%
    rename.('taxon' = genome)
emap$taxon %>% unique_n('genomes')
emap
```

#### COG

```r
# splitting COG categories
n_terms = emap$COG_category %>% sapply(stringr::str_length) %>% max
emap_f = emap %>%
    select.(taxon, query, COG_category) %>%
    separate.(COG_category, LETTERS[1:n_terms], sep='') %>%
    pivot_longer.(cols=c(-taxon, -query), 
                  names_to='X', values_to='COG_category') %>%
    filter.(!is.na(COG_category),
            COG_category != '-') %>%
    summarize.(n_genes = query %>% unique %>% length,
               .by = c(taxon, COG_category)) %>%
    mutate.(COG_category = gsub('^', 'num_genes_COG-', COG_category)) %>%
    pivot_wider.(names_from=COG_category, values_from=n_genes) %>%
    right_join(meta, c('taxon'='Sample')) %>%
    mutate.(across.(where(is.numeric), ~replace_na.(., 0)))

emap_f$taxon %>% unique_n('genomes')
emap_f
```

```r
# adding to db
db = db %>% 
    inner_join.(emap_f, c('Genome'='taxon', 'species'))
emap_f = NULL; gc()
db
```

#### KEGG pathways

```r
# splitting by KEGG pathways
n_terms = emap$KEGG_Pathway %>% sapply(function(x) stringr::str_count(x, ',')) %>% max
terms = gsub('^', 'X', 1:(n_terms+1))
perc_genomes = (meta$Sample %>% unique %>% length) * 0.7
                                       
emap_f = emap %>%
    select.(taxon, query, KEGG_Pathway) %>%
    separate.(KEGG_Pathway, terms, sep=',') %>%
    pivot_longer.(cols=c(-taxon, -query), 
                  names_to='X', values_to='KEGG_Pathway') %>%
    filter.(!is.na(KEGG_Pathway),
            grepl('^map', KEGG_Pathway)) %>%
    summarize.(n_genes = query %>% unique %>% length,
               .by = c(taxon, KEGG_Pathway)) %>%
    mutate.(KEGG_Pathway = gsub('^', 'num_genes_KEGG-', KEGG_Pathway)) %>%
    mutate.(n_genomes = taxon %>% unique %>% length,
            .by = c(KEGG_Pathway)) %>%
    filter.(n_genomes >= perc_genomes) %>%
    select.(-n_genomes) %>%
    pivot_wider.(names_from=KEGG_Pathway, values_from=n_genes) %>%
    right_join(meta, c('taxon'='Sample')) %>%
    mutate.(across.(where(is.numeric), ~replace_na.(., 0)))

emap_f$taxon %>% unique_n('genomes')
emap_f
```

```r
# adding to db
db = db %>% 
    inner_join.(emap_f, c('Genome'='taxon', 'species'))
emap_f = NULL; gc()
db
```

#### KEGG orthologs

```r
# splitting by KEGG orthologs
n_terms = emap$KEGG_ko %>% sapply(function(x) stringr::str_count(x, ',')) %>% max
terms = gsub('^', 'X', 1:(n_terms+1))
perc_genomes = (meta$Sample %>% unique %>% length) * 0.8
                                       
emap_f = emap %>%
    select.(taxon, query, KEGG_ko) %>%
    separate.(KEGG_ko, terms, sep=',') %>%
    pivot_longer.(cols=c(-taxon, -query), 
                  names_to='X', values_to='KEGG_ko') %>%
    filter.(!is.na(KEGG_ko)) %>%
    summarize.(n_genes = query %>% unique %>% length,
               .by = c(taxon, KEGG_ko)) %>%
    mutate.(KEGG_ko = gsub('^ko:', '', KEGG_ko)) %>%
    mutate.(KEGG_ko = gsub('^', 'num_genes_KO-', KEGG_ko)) %>%
    mutate.(n_genomes = taxon %>% unique %>% length,
            .by = c(KEGG_ko)) %>%
    filter.(n_genomes >= perc_genomes) %>%
    select.(-n_genomes) %>%
    pivot_wider.(names_from=KEGG_ko, values_from=n_genes) %>%
    right_join(meta, c('taxon'='Sample')) %>%
    mutate.(across.(where(is.numeric), ~replace_na.(., 0)))

emap_f$taxon %>% unique_n('genomes')
emap_f
```
